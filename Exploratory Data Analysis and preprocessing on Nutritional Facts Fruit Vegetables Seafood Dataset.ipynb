### Handling missing data

First, we need to understand the extent and pattern of missing data in our dataset. This can be done by checking for the presence of null or NaN values in our dataset.

```python
number_missing = dframe.isnull().sum()
percent_missing = dframe.isnull().sum() * 100 / len(dframe)
missing_value_dframe = pd.DataFrame({'number_missing':number_missing,'percent_missing': percent_missing})
missing_value_dframe
```

**Imputation**

Filling in the missing values with estimated or calculated values of median imputation. This helps to retain the information from other variables in the dataset.

```python
median_Dietary_Fiber = dframe["Dietary Fiber (g)"].median()
median_Dietary_Fiber_1 = dframe["Dietary Fiber.1 (%DV)"].median()
median_Sugars = dframe["Sugars (g)"].median()
median_Saturated_Fat = dframe["Saturated Fat (%DV)"].median()
median_Saturated_Fat_1 = dframe["Saturated Fat.1 (mg)Ê"].median()
median_Cholesterol = dframe["Chole-sterol (%DV)"].median()
median_Cholesterol = dframe["Chole-sterol.1 (mg)Ê"].median()

dframe = dframe.fillna({"Dietary Fiber (g)":median_Dietary_Fiber,"Dietary Fiber.1 (%DV)":median_Dietary_Fiber_1, 
                        "Sugars (g)":median_Sugars, "Saturated Fat (%DV)":median_Saturated_Fat,
                            "Saturated Fat.1 (mg)Ê":median_Saturated_Fat_1,"Chole-sterol (%DV)":median_Cholesterol, 
                        "Chole-sterol.1 (mg)Ê":median_Cholesterol})
```

```python
number_missing = dframe.isnull().sum()
percent_missing = dframe.isnull().sum() * 100 / len(dframe)
missing_value_dframe = pd.DataFrame({'number_missing':number_missing,'percent_missing': percent_missing})
missing_value_dframe
```
